```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

This shows us the sensors which are available and their dates

```python
url = "https://jonnyhigham.com/ENVS666/get_forecast.php"
response = requests.get(url)
data = response.json()

sensors = data['sensorInfo']
df = pd.DataFrame(sensors)

df
```

```python
startDate = '2024-01-01'
endDate = '2024-01-31'
sensor = 10

url = "https://jonnyhigham.com/ENVS666/get_forecast.php?sensor="+str(sensor)+"&startDate="+startDate+"&endDate="+endDate
response = requests.get(url)
data = response.json()

sensorInfo = data['sensorInfo']
data = data['data']
df = pd.DataFrame(data)

df
```

```python
time = pd.to_datetime(df['Timestamp'].array)
pollutant = df['PM2.5']

plt.plot(time,pollutant)
plt.xlabel('Time') # add some lables
plt.ylabel('Pollutant') # add some lables
```
